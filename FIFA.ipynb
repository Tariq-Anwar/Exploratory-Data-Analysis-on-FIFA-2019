3. Effect of Age on player's wage

```python
Mean_wage_per_age=df.groupby('age')['wage_eur'].mean()
```

```python
 Mean_wage_per_age
```

```python
plt=sns.barplot(x= Mean_wage_per_age.index,y=Mean_wage_per_age.values)
```

```python
talents_df=df.filter(["short_name","club","overall"])
talents_df = talents_df[talents_df.overall>=85]
talents_df = talents_df.groupby("club").count()
talents_df = talents_df.sort_values(by=['short_name'],ascending=False)
top_ten_clubs = talents_df[:10]
top_ten_clubs = list(top_ten_clubs.index.values)
top_count =  list(talents_df.iloc[:, 1])
```

```python
top_ten_clubs 
```

```python
top_ten_clubs_data = df.loc[df['club'].isin(top_ten_clubs), :]
```

4. Wages budget compare across top clubs

FC Barcelona and Real Madrid have the highest wage budgets, followed closely by Juventus and Manchester City

```python
ax = sns.barplot(x = "club",y = "wage_eur",data = top_ten_clubs_data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40)
```

5. Market values of players compare across top clubs. 

```python
sns.set(style="white")
p = sns.boxplot(x = 'club', y = 'value_eur', data = top_ten_clubs_data)
```

Spain, Brazil and Argentina have the players with highest overall scores

```python
top_ten_countries = df['nationality'].value_counts().head(10).index.values
top_ten_countries_data = df.loc[df['nationality'].isin(top_ten_countries), :]
p = sns.boxplot(x = 'nationality', y = 'overall', data = top_ten_countries_data)
```
